```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
from array import *
import matplotlib.mlab as mlab

mean = 2 # for uniform distribution in between [0,4]  E[X]=2
var = 2 # for uniform distribution in between [0,4] Var[X]=2
```

```python
def clt(k,n):
    a=[]
    for i in xrange(0,4*n+1):
        a.append(0);
    
    for j in range (1,k+1):
        w_n=0
        for i in range (1,n+1):
            w_n += random.randrange(0,5)
        a[w_n] += 1.0*sqrt(n*var)/k
    
    return a
```

```python
k=10000
for n in range(1,5):
  x=(linspace(0,4*n,4*n+1)-n*mean)/sqrt(n*var)
  y=clt(k,n)
  plt.subplot(2,2,n)
  z=arange(-5.,5,.1);
  plt.plot(z,mlab.normpdf(z,0,1))
  plt.plot(x,y,'ro')
#savefig('clt.png')
```
