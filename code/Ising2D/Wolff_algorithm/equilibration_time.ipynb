```python
Tc= 2.269
Temp = Tc 
J = 1.0 / Temp
H= 0.0
L = 16
N=L*L
Nobs = 4
Nsw=100
filename='equil'
```

```python
obs=np.zeros([Nobs, Nsw])
    
k = open('%s/observable_J%.5f_L%i_H%.2f' %(filename , J, L,H), 'r')
    
for a in xrange(0,Nsw):
    for i in xrange(0,Nobs):
        obs[i][a]=k.readline()
```

```python
x=linspace(0,Nsw-1,Nsw)
plt.plot(x,obs[0])
plt.title('Magnetization per Site(m)',fontsize=12)
plt.xlabel('MC Time per Site',fontsize=12)
#plt.text(310, 0.2, r' m =$ \frac{1}{N} \sum s_i$',fontsize=18)
#plt.xlim(0,45)
#plt.ylim(-1,0.01)
#savefig('/Users/onurkaplan/desktop/grafikler/equilibration_time/magnet_allup_T%f.png' %Temp )
plt.show()
```
