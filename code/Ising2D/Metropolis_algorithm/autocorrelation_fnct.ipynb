```python
j=0
log_auto=[]

while auto[j] > 0 :        
    log_auto.append(np.log(auto[j]))
    j = j+1
```

```python
gg=20
a = np.linspace(0,len(log_auto)-1-gg,len(log_auto)-gg)
plt.plot(a,log_auto[0:len(log_auto)-gg])
```

```python
p = np.polyfit(a,log_auto[0:len(log_auto)-gg],1)
print p 
```
