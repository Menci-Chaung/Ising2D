```python
Temp= 3.0
J = 1 / Temp
H= 0.0
L = 32
N=L*L
Nobs = 4
Nsw=3000
filename='equil'
```

```python
obs=np.zeros([Nobs, Nsw])
    
k = open('%s/observable_J%.2f_L%i_H%.2f' %(filename , J, L,H), 'r')
    
for a in xrange(0,Nsw):
    for i in xrange(0,Nobs):
        obs[i][a]=k.readline()
```

```python
print len(obs[0])
```

```python
x=linspace(0,Nsw-1,Nsw)
plt.plot(x,obs[0])
plt.title('Magnetization per Site(m)',fontsize=12)
plt.xlabel('MC Time per Site',fontsize=12)
plt.text(500, 0.6, r' m =$ \frac{1}{N} \sum s_i$',fontsize=18)
plt.xlim(0,800)
#plt.ylim(0,1)
savefig('../../graphics/equilibration_time/magnet_allup_T%f_L%i.png' %(Temp,L) )
plt.show()
```
