```python
L=[6,9,12,24,48]
corr=[113,309,568,2308,7800]
```

```python
logL= np.log(L)
logcorr=np.log(corr)
```

```python
p = np.polyfit(logL,logcorr,1)
print p 
```

```python
plt.plot(logL,logcorr,'o')
plt.plot(logL,p[0]*logL+p[1],'r')
plt.title(r'$Log\tau $',fontsize=12)
plt.xlabel('$logL$',fontsize=12)
plt.text(3, 5.5, r' z = %f' %p[0],fontsize=12)
#plt.xlim(0,45)
#plt.ylim(0.2,4.07)
savefig('/Users/onurkaplan/dropbox/kaplan/programs/graphics/properties/dynamic_exponent.png')
plt.show()
```
